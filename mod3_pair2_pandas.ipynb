```python
# Extraed las filas pares, de las columnas "rating", "rating_count", "rating_five_count", "rating_four_count", "rating_three_count", "rating_two_count", "rating_one_count".

columnas_deseadas = [
    "rating",
    "rating_count",
    "rating_five_count",
    "rating_four_count",
    "rating_three_count",
    "rating_two_count",
    "rating_one_count"
]

# Extraer filas pares (índices pares) y columnas seleccionadas
filas_pares = df_comercio.loc[df_comercio.index % 2 == 0, columnas_deseadas]
print(filas_pares)
```
